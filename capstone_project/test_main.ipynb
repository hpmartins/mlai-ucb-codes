```python
import pandas as pd

from nba_api.stats import endpoints

pd.set_option('display.max_columns', None)
pd.set_option('display.min_rows', 12)
pd.set_option('display.max_rows', 100)
```

```python
players = pd.DataFrame(endpoints.PlayerIndex().get_data_frames()[0])
players.sort_values('PTS', ascending = False)

teams = pd.DataFrame(endpoints.CommonTeamYears().get_data_frames()[0])
teams.MAX_YEAR = pd.to_numeric(teams.MAX_YEAR)
teams = teams[teams.MAX_YEAR == 2023].reset_index(drop=True)

def get_team_id(team):
    return teams[teams.ABBREVIATION == team].TEAM_ID.values[0]

def get_player_id(player):
    return players[players.PLAYER_SLUG == player].PERSON_ID.values[0]

def get_game_logs(team):
    return pd.DataFrame(endpoints.LeagueGameFinder(team_id_nullable = get_team_id(team)).get_data_frames()[0])
```
